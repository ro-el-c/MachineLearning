```
import numpy as np
import torch

# 데이터 읽기 함수
def load_dataset(file):

  data = np.loadtxt(file)
  print("DATA=",data)
  
  x = data[:,0:-1] # , 앞은 row에 대하여 / , 뒤는 column
  print("X=",x)
  
  y = np.reshape(data[:,-1],(3,1))
  print("Y=",y)
 
  # 텐서로 변경
  x = torch.tensor(x, dtype=torch.float)
  y = torch.tensor(y, dtype=torch.float)

  return (x, y)


# 텐서를 리스트로 바꾸는 함수
# 텐서 -> numpy -> list
def tensor2list(input_tensor):
  return input_tensor.cpu().detach().numpy().tolist()

x, y = load_dataset("/gdrive/My Drive/ml_colab/week2/train.txt")

# 가중치 초기화
w = torch.randn(1)*3.0
print("\nInit: w = {0}\n".format(tensor2list(w)))

# 학습률
alpha = 0.1

# 학습데이터 수
m = y.size(0)

# 1 epoch ; 데이터 전체를 한 번 다 읽었을 때
## 아래 for문은 w 업데이트를 50번

# epoch batch step
# epoch: 데이터 전체를 한 번 다 읽었을 때
# -> 여러 개의 batch로 나눔
#            -> 하나의 batch = 1 step
# ex) epoch 1000개 => 100개씩 가져와 batch 구성 * 10 step
# 이해를 위한 수식) epoch = batch * step

for epoch in range(50):
  
  cost=0

  for i in range(m):
    cost += (w * x[i] - y[i]) * x[i]

  cost = 1/m * cost
  w -= alpha * cost
  
  print("Epoch {0:d}: w = {1}".format(epoch+1,tensor2list(w)))
```
