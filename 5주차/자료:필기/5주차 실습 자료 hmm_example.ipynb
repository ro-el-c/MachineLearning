# transition prob 만 계산 == MM

# observation prob & transition prob == HMM

initial prob (파이) - 첫 시작 확률 이미 정해져 있는 확률

```
import numpy as np
from hmmlearn import hmm

states = ["Rainy", "Cloudy", "Sunny"]
n_states = len(states)

observations = ["Boots", "Shoes"]
n_observations = len(observations)

# 시작 확률 - pi
start_probability = np.array([0.2, 0.5, 0.3]) # R, C, S

# 전이 확률
# -> 원래는 training data로 부터 구해야 함
# rainy 다음 날 rainy 인 observation probability -> rainy인 경우 전부 구한 후, rainy-rainy인 경우 구하기 (MLE)
# 곱에 log 씌우면 합
# (값이 0이 안 되도록 하려고 스무딩. -> 너무 큰 애는 좀 줄이고, 너무 작은 애는 좀 늘리고) log likelyhood estimation
transition_probability = np.array([
  # R    C    S
  [0.4, 0.3, 0.3], # R
  [0.2, 0.6, 0.2], # C
  [0.1, 0.1, 0.8] # S
])

# 관측 확률 - state에서 나올 확률
# = observation probability
# 왜 emission probability???????
emission_probability = np.array([
  # B    S
  [0.8, 0.2], # R
  [0.5, 0.5], # C
  [0.1, 0.9] # S
])

# 모델 만들기
# model = hmm.MultinomialHMM(n_components=n_states)
model = hmm.CategoricalHMM(n_components=n_states)
# 여기를 채우세요.
model.startprob_ = start_probability
model.transmat_ = transition_probability
model.emissionprob_ = emission_probability

# 관측 입력
input = [0, 0, 1, 1]

# HMM 모델 호출
# 여기를 채우세요.
hmm_input = np.atleast_2d(input).T
logprob, sequence = model.decode(hmm_input)
# tuple 형태로 리턴.

# observation*transition -> 하나라도 0이 되면 모두 0이 됨
# 곱에 log 씌워서 합으로 바꿈. ex) log(a*b) = log(a) + log(b)


# input이 리스트면, 하나씩 가져와서 x에 넣어줌 -> 그 결과를 람다가 출력 -> join이 모두 합침
print("Input :", ", ".join(map(lambda x: observations[x], input)))
print("Output:", ", ".join(map(lambda x: states[x], sequence)))
print("Prob. :", logprob)
# logprob 값: - 값이 보기 싫으면 다시 - 곱해주면 됨 -> negative log likelyhood => 확률이 클수록 작아짐(로스에 많이 쓰임)
```
