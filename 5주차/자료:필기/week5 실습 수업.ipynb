## HMM(Hidden Markov Model)

**Markov Model**: 상태(정해져 있음)가 있고, **상태 간 전이되는 확률값**을 채워넣은 것

joint prob -> chain rule -> 1차 Markov Assumption
1차 Markov Model == Bigram Model

<br>

MM <-> HMM

가장 큰 차이: state가 결정되어 있지 않다. observation한 것에 대한 확률값으로 존재

1차 Markov Assumption + Independent Assumption (확률식 유도할 줄 알아야 함)

<br>

**목적**: 최적의 state 열을 찾는 것

- 가능한 모든 combination을 해보는 것
