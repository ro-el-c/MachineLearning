# 굉장히 중요

### 층을 계속 늘리면 안되는 이유
-> Vanish Gradient 문제
<br>
:F(F(F(F(F(...F(x))...))))
0과 1 사이 값을 계속 곱해져 나가면 결국 0과도 같아지기 때문

<br>
### 해결
Activation Function -> ReLu
<br>미분하면 1 -> 계속 1 곱해나감
<br>Dropout: 노이즈 줘서 일부러 끊어 나감 : 일반화 성능 높이기 위한 시도

<br>
<br>
Residual Connection
<br>: 우회하여 상위층으로 연결

<br>
<br>
층을 높일수록 추상화가 많이 됨 -> 원본의 의미를 잃음
<br>
<br>
Underfitting: 학습 더 시키기
<br>OverFitting: 전체 weight가 너무 세세 / 학습 덜 시키기
<br>-> L1 정규화: 절댓값의 함 / L2 정규화: 제곱의 합(거리)
