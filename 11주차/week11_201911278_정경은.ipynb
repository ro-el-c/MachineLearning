## 한국어 Word2Vec을 이용하여 아래에 주어진 단어들을 벡터 연산한 후에 가장 가까운 상위 3개의 결과를 제시하시오.

- Word2Vec: https://github.com/Kyubyong/wordvectors
- 벡터 연산
    - 한국-서울+파리
    - 컴퓨터+공학+인문학
    - 사랑+이별
    - 충청북도-청주+강원도
    - 인간-사랑
    - 미술-물감+음악

방법에 따라 2가지 결과가 나올 수 있음
<br>
이 중 **벡터 연산**을 직접해서 얻은 결과를 인정

```python
from gensim.models import Word2Vec

# Word2Vec 모델 로드하기
loaded_model = Word2Vec.load('/gdrive/My Drive/ml_colab/week11/data/ko.bin')
```

```python
# 1. 한국-서울+파리
korea = loaded_model.wv['한국']
seoul = loaded_model.wv['서울']
paris = loaded_model.wv['파리']

p1 = korea-seoul+paris
#print(p1)


# 2. 컴퓨터+공학+인문학
computer = loaded_model.wv['컴퓨터']
engineering = loaded_model.wv['공학']
humanities = loaded_model.wv['인문학']

p2 = computer+engineering+humanities
#print(p2)


# 3. 사랑+이별
love = loaded_model.wv['사랑']
breakup = loaded_model.wv['이별']

p3 = love+breakup
#print(p3)


# 4. 충청북도-청주+강원도
chungcheongbukdo = loaded_model.wv['충청북도']
cheongju = loaded_model.wv['청주']
gangwondo = loaded_model.wv['강원도']

p4 = chungcheongbukdo-cheongju+gangwondo
#print(p4)


# 5. 인간-사랑
human = loaded_model.wv['인간']

p5 = human+love
#print(p5)


# 6. 미술-물감+음악
art = loaded_model.wv['미술']
paint = loaded_model.wv['물감']
music = loaded_model.wv['음악']

p6 = art-paint+music
#print(p6)

print("results:")
print(loaded_model.wv.most_similar([p1], topn=3))
print(loaded_model.wv.most_similar([p2], topn=3))
print(loaded_model.wv.most_similar([p3], topn=3))
print(loaded_model.wv.most_similar([p4], topn=3))
print(loaded_model.wv.most_similar([p5], topn=3))
print(loaded_model.wv.most_similar([p6], topn=3))
```
